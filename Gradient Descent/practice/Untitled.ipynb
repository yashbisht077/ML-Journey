```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {"age":[5,15,52,25,42],"salary":[15000,5000,30000,20000,11000]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.scatter("age","salary",data = df)
plt.show()
```

```python
x = df[["age"]]
y = df["salary"]
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
y_predict = lr.predict(x)
```

```python
lr.coef_,lr.intercept_
```

```python
m = 1
```

```python
c = 20000
```

```python
y_p = ( df["age"] * m ) + c
```

```python
plt.scatter("age","salary",data = df)
plt.plot(x,lr.predict(x))
plt.show()
```

```python
plt.scatter("age","salary",data = df)
plt.plot(x,y_p)
plt.show()
```
