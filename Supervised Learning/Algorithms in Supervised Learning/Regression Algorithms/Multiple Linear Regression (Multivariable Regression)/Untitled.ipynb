```python
# Import necessary libraries
from sklearn.model_selection import train_test_split  # For splitting data into training and testing sets
from sklearn.linear_model import LinearRegression    # For implementing linear regression
import pandas as pd                                   # For data manipulation
import numpy as np                                    # For numerical operations

# For data visualization
import plotly.express as px                           # For creating interactive 3D scatter plots
import plotly.graph_objects as go                    # For advanced plot customization

# For evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score  # For performance evaluation
```

```python
dataset = pd.read_csv(r"Data/SalaryMulti.csv",usecols=["Total Experience","Team Lead Experience","Project Manager Experience","Salary"])
dataset.head()
```

```python
dataset.corr()
```

```python
fig = px.scatter_3d(dataset,x = "Total Experience",y="Team Lead Experience",z="Salary")
# fig.write_html("3d_plot.html")
fig.show()
```

```python
X = dataset[["Total Experience","Team Lead Experience","Project Manager Experience"]]
y = dataset["Salary"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
# Evaluate the model using MAE, MSE, and R² metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R² Score:", r2)
```
