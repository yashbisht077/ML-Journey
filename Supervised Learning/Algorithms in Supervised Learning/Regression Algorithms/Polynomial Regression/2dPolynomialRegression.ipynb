```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Generate synthetic data
np.random.seed(42)
x = np.linspace(-3, 3, 100).reshape(-1, 1)  # Independent variable
y = 2 + 3*x + 0.5*x**2 + np.random.randn(100, 1) * 2  # Quadratic relationship with noise
```

```python
# Split the data into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Training data points:", len(x_train))
print("Testing data points:", len(x_test))
```

```python
# Plot the original data points
plt.scatter(x, y, color='blue', label='Original Data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Data Visualization')
plt.legend()
plt.show()
```

```python
# Define the degree of the polynomial
degree = 2

# Create polynomial features for the training and testing sets
poly = PolynomialFeatures(degree=degree)
X_train_poly = poly.fit_transform(x_train)
X_test_poly = poly.transform(x_test)
```

```python
# Fit a linear regression model to the training polynomial features
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Make predictions on training and testing sets
y_train_pred = model.predict(X_train_poly)
y_test_pred = model.predict(X_test_poly)
```

```python
# Compute Mean Squared Error for training and testing sets
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print("Mean Squared Error on Training Set:", mse_train)
print("Mean Squared Error on Testing Set:", mse_test)
```

```python
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.plot(np.sort(x_train, axis=0), np.sort(y_train_pred, axis=0), color='red', label='Polynomial Fit (Train)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression - Training Set')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test, y_test, color='green', label='Testing Data')
plt.plot(np.sort(x_test, axis=0), np.sort(y_test_pred, axis=0), color='orange', label='Polynomial Fit (Test)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression - Testing Set')
plt.legend()
plt.show()
```
